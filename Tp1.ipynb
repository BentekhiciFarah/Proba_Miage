### Solution 2 ( Brute force )

A)  Quelle est la probabilité que les deux premières cartes distribuées au joueur
 totalisent 21 points (blackjack) ?

```python
Cartes = ['as', '2', '3', '4', '5', '6', '7', '8', '9', '10', 'R', 'D', 'V'] # Créer les rangs
CARTES = [coul for coul in Cartes for _ in range(4)] # Créer 4 cartes de chaque rang
print(CARTES)
```

```python
Cartes_rep = pd.DataFrame(itertools.combinations(CARTES, 2)) # Lister tous les cas possibles et les mettre dans un tableau
Cartes_rep.columns = Cartes_rep. columns + 1 #Avancer l'indice de 0 à 1
Cartes_rep # Afficher toutes les possibilités
```

```python
res = Cartes_rep[(Cartes_rep[1] == 'as') &  (Cartes_rep[2].isin(['D', 'V', 'R', '10']))]
print(res)
```

```python
print("La probabilité de gagner dès les deux premières cartes est de : ",len(res) / len(Cartes_rep))
```

### Solution 3 Probabiliste

```python
SEED = 1234

RNG = np.random.default_rng(SEED)   #Crée un générateur de nbr aléatoires
```

```python
def rdm_exp(rng, Cartes):
    return rng.permutation(Cartes).tolist()         #Créee une liste de cartes
```

```python
rdm_exp(RNG, CARTES)                    # Mélanger les cartes
```

```python
def rdm_rep(rng, cartes, nb_rep) :          #Fonction créant nb_rep jeux de cartes
    return [rdm_exp(RNG, cartes) for _ in range(nb_rep)]
```

```python
nb_rep = 100000
comb = pd.DataFrame(rdm_rep(RNG, CARTES, nb_rep))       #Transformer en DataFrame
comb
```

```python
comb_2 = comb.iloc[:, :2]                   #Tirer 2 cartes de chaque jeu de 52
comb_2.columns = comb_2.columns + 1
comb_2
```

```python
res = comb_2[(comb_2[1] == 'as') &  (comb_2[2].isin(['D', 'V', 'R', '10']))]
res
```

```python
print("Le nbr de cas correspondant est de : ", len(res))
print("La probabilité estimée est de : ", len(res) / nb_rep)
```
